```python
import pandas as pd
```

```python
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTKv5ve5cbcTUOygS_FEd-swTOTVxC4DldFq-T2W4nPW-XqMYK8rVrMn_RDwf_ujF8o_1YyHAL98OiO/pub?gid=1414694331&single=true&output=csv")
```

```python
pd.set_option("display.width",200)
#print(df)
df
```

```python
df_juliaca = df[df["UBIGEO"] == 211101]
print("Tenemos",len(df_juliaca), "registros en Juliaca.")

df_puno = df[df["UBIGEO"] == 210101]
print("Tenemos",len(df_puno), "registros en Puno.")

df_puno_total = df[df["UBIGEO"].astype(str).str.startswith('21')]
print("Tenemos",len(df_puno_total), "registros en total para todo Puno.")

df_tacna = df[df["UBIGEO"] == 230102]
print("Tenemos",len(df_tacna), "registros en Tacna.")
```

```python
import matplotlib.pyplot as plt
```

```python
# Contar registros para cada ubicación
juliaca_count = len(df_juliaca)
puno_count = len(df_puno)
puno_total_count = len(df_puno_total)
tacna_count = len(df_tacna)

# Etiquetas y valores para el gráfico de pastel
labels = ['Juliaca', 'Puno', 'Total Puno', 'Tacna']
sizes = [juliaca_count, puno_count, puno_total_count, tacna_count]

# Crear el gráfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribución de Registros por Ubicación')
plt.show()
```

```python
#filtrar los registros de Puno y Juliaca
df_puno_juliaca = df[(df["UBIGEO"]==210101) | (df["UBIGEO"]==211101)]
print(df_puno_juliaca)
```

```python
#Obtenemos el numero de damas y varones  en Puno y Juliaca de acuerdo a la columna "sexo"

# Varones de Puno
varones_puno = df_puno[df_puno["sexo"]==1.0]
print("Cantidad de varones: ",len(varones_puno))

# Damas de Puno
damas_puno = df_puno[df_puno["sexo"]==2.0]
print("Cantidad de damas: ",len(damas_puno))

# Varones de Juliaca
varones_juliaca = df_juliaca[df_juliaca["sexo"]==1.0]
print("Cantidad de varones: ",len(varones_juliaca))

# Damas de Juliaca
damas_juliaca = df_juliaca[df_juliaca["sexo"]==2.0]
print("Cantidad de damas: ",len(damas_juliaca))
```

```python
import matplotlib.pyplot as plt

# Contar la cantidad de varones y damas en Puno y Juliaca
varones_puno_count = len(varones_puno)
damas_puno_count = len(damas_puno)
varones_juliaca_count = len(varones_juliaca)
damas_juliaca_count = len(damas_juliaca)

# Etiquetas para el gráfico de pastel
labels_pie = ['Varones Puno', 'Damas Puno', 'Varones Juliaca', 'Damas Juliaca']
# Valores para el gráfico de pastel
sizes_pie = [varones_puno_count, damas_puno_count, varones_juliaca_count, damas_juliaca_count]
# Colores para el gráfico de pastel
colors_pie = ['lightblue', 'lightcoral', 'lightskyblue', 'lightpink']
# Separación de la porción más grande
explode = (0.1, 0, 0.1, 0)

# Crear el gráfico de pastel
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
patches, texts, autotexts = plt.pie(sizes_pie, explode=explode, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Género en Puno y Juliaca')

# Añadir las cantidades al costado de los porcentajes
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')
    autotext.set_horizontalalignment('center')
    autotext.set_verticalalignment('center')
    autotext.set_text(autotext.get_text() + f"\n({sizes_pie[autotexts.index(autotext)]})")

# Etiquetas y alturas para el gráfico de barras
labels_bar = ['Puno', 'Juliaca']
varones_counts = [varones_puno_count, varones_juliaca_count]
damas_counts = [damas_puno_count, damas_juliaca_count]

# Crear el gráfico de barras
plt.subplot(1, 2, 2)
bar_width = 0.35
r1 = range(len(labels_bar))
r2 = [x + bar_width for x in r1]
plt.bar(r1, varones_counts, color='lightblue', width=bar_width, label='Varones')
plt.bar(r2, damas_counts, color='lightcoral', width=bar_width, label='Damas')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad')
plt.title('Cantidad de Varones y Damas en Puno y Juliaca')
plt.xticks([r + bar_width/2 for r in range(len(labels_bar))], labels_bar)
plt.legend()

# Ajustar el diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
```

```python
df_infantes = df_puno_total[(df_puno_total["edad"]>3) & (df_puno_total["edad"]<12)]
print(len(df_infantes))
```

```python
df_infantes_quechua_varon = df_infantes[(df_infantes["idioma_ninez"]==1) & (df_infantes["sexo"]==1.0)]
print("Cantidad de niños que hablan quechua: ",len(df_infantes_quechua_varon))

df_infantes_quechua_dama = df_infantes[(df_infantes["idioma_ninez"]==1) & (df_infantes["sexo"]==2.0)]
print("Cantidad de niñas que hablan quechua: ",len(df_infantes_quechua_dama))

df_infantes_aymara_varon = df_infantes[(df_infantes["idioma_ninez"]==2) & (df_infantes["sexo"]==1.0)]
print("Cantidad de niños que hablan aymara: ",len(df_infantes_aymara_varon))

df_infantes_aymara_dama = df_infantes[(df_infantes["idioma_ninez"]==2) & (df_infantes["sexo"]==2.0)]
print("Cantidad de niñas que hablan aymara: ",len(df_infantes_aymara_dama))
```

```python
import matplotlib.pyplot as plt

# Cantidades de niños y niñas que hablan Quechua y Aymara
niños_quechua = len(df_infantes_quechua_varon)
niñas_quechua = len(df_infantes_quechua_dama)
niños_aymara = len(df_infantes_aymara_varon)
niñas_aymara = len(df_infantes_aymara_dama)

# Crear figuras y subgráficos
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico de pastel
labels = ['Niños que hablan Quechua', 'Niñas que hablan Quechua', 'Niños que hablan Aymara', 'Niñas que hablan Aymara']
sizes = [niños_quechua, niñas_quechua, niños_aymara, niñas_aymara]
explode = (0.1, 0, 0.1, 0)
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

wedges, texts, autotexts = axs[0].pie(sizes, explode=explode, labels=None, autopct='%1.1f%%', shadow=True, startangle=140, colors=colors, wedgeprops=dict(width=0.4))

for i, (text, autotext) in enumerate(zip(texts, autotexts)):
    percentage = sizes[i] / sum(sizes) * 100
    text.set_text(f'{labels[i]}\n({sizes[i]})')
    autotext.set_text(f'{percentage:.1f}%')
    autotext.set_color('black')
    autotext.set_fontsize(10)

axs[0].set_title('Distribución de Infantes por Idioma y Género')
axs[0].legend(wedges, labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Gráfico de barras
idiomas = ['Quechua', 'Aymara']
niños = [niños_quechua, niños_aymara]
niñas = [niñas_quechua, niñas_aymara]
x = range(len(idiomas))
width = 0.35

axs[1].bar(x, niños, width, label='Niños', color='#66b3ff')
axs[1].bar(x, niñas, width, bottom=niños, label='Niñas', color='#ff9999')
axs[1].set_ylabel('Cantidad')
axs[1].set_title('Distribución de Infantes por Idioma y Género')
axs[1].set_xticks(x)
axs[1].set_xticklabels(idiomas)
axs[1].legend()

# Ajustar diseño y mostrar gráficos
plt.tight_layout()
plt.show()
```

```python
# Load the CSV files from the provided URLs
df1 = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQNGmvL3cPqc_h1MiWwu5YOFtozYuleNuZR9gAS8xgBiMoY5dNpfwGVOBmsMEqRIfYpJac2YjsOUI4K/pub?gid=1859822377&single=true&output=csv")
df2 = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQZNvhw7mFKTTWwmZm0gmgW3nZ8T0W4NO5tkK7zCV8dQoe2WVK1xXUGSEuY6xwM6JrNB-RO6kLJYzOM/pub?gid=1185186065&single=true&output=csv")
```

```python

# ... Rest of your code for creating charts using df1 and df2 ...

plt.figure(figsize=(12, 6))
plt.bar(df1["anual"], df1["produccion"], label="DataFrame1", width=0.4, color="blue")
plt.bar(df2["anual"] + 0.4, df2["produccion"], label="DataFrame2", width=0.4, color="red")
# ... rest of the chart code ...


```

```python
common_columns = set(df1.columns) & set(df2.columns)

# Exclude unnecessary columns (e.g., "anual" if it's just for labeling)
common_columns = common_columns.difference(["anual"])  # Modify as needed

for col in common_columns:
  plt.figure(figsize=(10, 6))
  plt.bar(df1["anual"], df1[col], label="DataFrame1", width=0.4, color="blue")
  plt.bar(df2["anual"] + 0.4, df2[col], label="DataFrame2", width=0.4, color="red")
  plt.xlabel("Año (solo para barras)")  # Adjust label if not using bars
  plt.ylabel(col)
  plt.title(f"Comparación de {col}")
  plt.xticks(rotation=45)  # Optional: Rotate x-axis labels for better readability
  plt.legend()
  plt.tight_layout()
  plt.show()
```
